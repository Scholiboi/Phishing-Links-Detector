```python
# Define a function to count stars, dashes, and underscores in a string
def count_special_chars(link):
    stars = link.count('*')
    dashes = link.count('-')
    underscores = link.count('_')
    dots = link.count('.')

    return pd.Series([stars, dashes, underscores, dots])

# Apply the function to the 'phish_detail_url' column and concatenate the results
df[['stars', 'dashes', 'underscores', 'dots']] = df['url'].apply(count_special_chars)

# Display the updated DataFrame
df
```

```python
import re
from urllib.parse import urlparse
import pandas as pd

df = pd.read_csv('verified_online.csv')
df.drop(['phish_detail_url'], axis="columns", inplace=True)

# import pandas as pd

# Define a function to extract features using brute force
def brute_force_extract_features(url):
    # Check for empty or invalid URLs
    if not url or not isinstance(url, str):
        return pd.Series([None] * 21)  # Return NaN for all features if URL is invalid

    # Basic string counting and checks
    num_dots = url.count('.')
    num_dash = url.count('-')
    num_underscore = url.count('_')
    num_percent = url.count('%')
    num_query_components = url.count('&')
    num_ampersand = num_query_components  # Same as query components
    num_hash = url.count('#')
    num_numeric_chars = sum(c.isdigit() for c in url)
    url_length = len(url)
    path_level = url.count('/')
    at_symbol = url.count('@')
    tilde_symbol = url.count('~')
    hostname_length = len(url.split('/')[0].split(':')[0])  # Assuming domain is before first '/' or ':'
    path_length = len(url.split('/')[1]) if len(url.split('/')) > 1 else 0  # Length after domain
    query_length = len(url.split('?')[1]) if '?' in url else 0  # Length of query part after '?'
    double_slash_in_path = url.count('//')  # Count instances of double slashes in path

    # Check if URL starts with 'https'
    no_https = int(not url.startswith('https'))

    # Simple heuristic to calculate subdomain level (dots in hostname before first '/')
    hostname = url.split('/')[0]  # Get hostname part before first '/'
    subdomain_level = hostname.count('.') - 1 if hostname else 0

    # Return all features as a pandas Series
    return pd.Series([num_dots, subdomain_level, path_level, url_length, num_dash, 
                      num_dash, at_symbol, tilde_symbol, num_underscore, num_percent, 
                      num_query_components, num_ampersand, num_hash, num_numeric_chars, 
                      no_https, hostname_length, path_length, query_length, double_slash_in_path])

# Apply the function to extract features
df[['NumDots', 'SubdomainLevel', 'PathLevel', 'UrlLength', 'NumDash', 
    'NumDashInHostname', 'AtSymbol', 'TildeSymbol', 'NumUnderscore', 'NumPercent', 
    'NumQueryComponents', 'NumAmpersand', 'NumHash', 'NumNumericChars', 'NoHttps', 
    'HostnameLength', 'PathLength', 'QueryLength', 'DoubleSlashInPath']] = df['url'].apply(brute_force_extract_features)

df
```
